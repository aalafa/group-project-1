```python
# Dependencies
import requests
import pandas as pd
from census import Census
import json
```

```python
# Import U.S. Census API Key
from config import census_key

# Create an instance of the Census library
c = Census(
    census_key,
    year = 2010
)
```

```python
# Run Census Search to retrieve data on all zip codes (2013 ACS5 Census)
census_data = c.acs5.get(
    (
        "NAME",
        "B19001_001E",
        "B19001A_001E",
        "B19001B_001E",
        "B19001C_001E",
        "B19001D_001E",
        "B19001E_001E",
        "B19001F_001E",
        "B19001G_001E",
        "B19001I_001E"
    ),
    {'for': 'state:*'}
)

# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column renaming
census_pd = census_pd.rename(
    columns = {
        "B19001_001E": "Household Income Total",
        "B19001A_001E": "White Alone Householder",
        "B19001B_001E": "African American Alone Householder",
        "B19001C_001E": "American Indian and Alaska Native Alone Householder",
        "B19001D_001E": "Asian Alone Householder",
        "B19001E_001E": "Native Hawaiian and Other Pacific Isalnder Alone Householder",
        "B19001F_001E": "Some Other Race Alone Householder",
        "B19001G_001E": "Two or more Races Householder",
        "B19001I_001E": "Hispanic or Latino Householder",
        "NAME": "Name"
    }
)

# Add a Poverty Rate column (Poverty Count / Population)
#census_pd["Poverty Rate"] = 100 * census_pd["Poverty Count"].astype(int) / census_pd["Population"].astype(int)

# Configure the final DataFrame
census_pd = census_pd[
    [
        "Name",
        "Household Income Total",
        "White Alone Householder",
        "African American Alone Householder",
        "American Indian and Alaska Native Alone Householder",
        "Asian Alone Householder",
        "Native Hawaiian and Other Pacific Isalnder Alone Householder",
        "Some Other Race Alone Householder",
        "Two or more Races Householder",
        "Hispanic or Latino Householder"
    ]
]

# Display DataFrame length and sample data
print(f"Number of rows in the DataFrame: {len(census_pd)}")
census_pd
```
